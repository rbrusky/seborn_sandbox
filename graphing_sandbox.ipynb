```python
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
import pandas as pd
```

```python
def format_percent(x):
    return float(str(x['% in wells']).split('%')[0])/100
```

```python
df_emily = pd.read_csv('Olfactory_Assays.csv')
dilution_sort = {'0':0, '10^-5':1, '10^-4':2, '10^-3':3, '10^-2':4, '10^-1':5, '1':6}
df_emily['dilution_order'] = df_emily['Dilution']
df_emily = df_emily.replace({'dilution_order': dilution_sort})
df_emily['percent'] = df_emily.apply(lambda x : format_percent(x), axis=1)
df_emily
```

```python
# Draw a pointplot to show well percent as a function of dilution factors
g = sns.factorplot(x="Dilution", y="percent", col="Compound", data=df_emily,
                   capsize=.2, palette="YlGnBu", size=6, aspect=.75)
g.despine(left=True)
```

```python
sns.set(font_scale=1)
sns.boxplot(x="Dilution", y="percent", data=df_emily, palette="pink_r")
sns.despine(offset=10, trim=True)
```

```python
sns.set(font_scale=1)
sns.boxplot(x="Dilution", y="percent", data=df_emily, palette="YlGnBu")
sns.despine(offset=10, trim=True)
```
